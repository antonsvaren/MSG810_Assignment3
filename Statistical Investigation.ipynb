```python
# basic sums
y_dot_dot_dot = df['Mass'].mean()
y_i_dot_dot = df.groupby('Color')['Mass'].mean()
y_dot_j_dot = df.groupby(['Store', 'Location', 'Number'])['Mass'].mean()
y_i_j_dot = df.groupby(['Store', 'Location', 'Number', 'Color'])['Mass'].mean()

# SS_A
balls_per_color = df.groupby('Color')['Mass'].count()
squares_A = (y_i_dot_dot - y_dot_dot_dot)**2
SS_A = (balls_per_color * squares_A).sum()

# SS_B
balls_per_bag = df.groupby(['Store', 'Location', 'Number'])['Mass'].count()
squares_B = (y_dot_j_dot - y_dot_dot_dot)**2
SS_B = (balls_per_bag * squares_B).sum()

# SS_AB
balls_per_bag_and_color = df.groupby(['Store', 'Location', 'Number', 'Color'])['Mass'].count()
squares_AB = (y_i_j_dot - y_i_dot_dot - y_dot_j_dot + y_dot_dot_dot)**2
SS_AB = (balls_per_bag_and_color * squares_AB).sum()

# SS_E
df_joined = df.join(y_i_j_dot, on=['Store', 'Location', 'Number', 'Color'], rsuffix='_other')
squares_E = (df_joined['Mass'] - df_joined['Mass_other']) ** 2
SS_E = squares_E.sum()

# dimensions
a = y_i_dot_dot.count()
b = y_dot_j_dot.count()

# degrees of freedom
a_deg_freedom = a-1
b_deg_freedom = b-1
ab_deg_freedom = (a-1) * (b-1)
e_deg_freedom = (balls_per_bag_and_color - 1).sum()

# mean squares
MS_A = SS_A / a_deg_freedom
MS_B = SS_B / b_deg_freedom
MS_AB = SS_AB / ab_deg_freedom
MS_E = SS_E / e_deg_freedom

# F-test statistics
F_A = MS_A/MS_E
F_B = MS_B/MS_E
F_AB = MS_AB/MS_E



print("\nSum of squares: ")

print(SS_A)
print(SS_B)
print(SS_AB)
print(SS_E)


print("\nDegrees of freedom: ")

print(a_deg_freedom)
print(b_deg_freedom)
print(ab_deg_freedom)
print(e_deg_freedom)


print("\nMean squares: ")

print(MS_A)
print(MS_B)
print(MS_AB)
print(MS_E)


print("\nF-test statistics: ")

print(F_A)
print(F_B)
print(F_AB)



print(balls_per_color)
```
