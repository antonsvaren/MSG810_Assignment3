## Introduction

### Problem description

<br> In this study, we have chosen to do a statistical analysis of the weights of M&M Peanut candy
balls. In specific, we set out to answer the following two questions: <br>
<br>
- Does the weight of an M&M ball depend on its color?
- Does the weight of an M&M ball depend on which bag it was taken from?
</ul>
While we did not expect to find any groundbreaking results from answering these questions,
the above problem formulation can be considered similar to the types of properties monitored
in a “quality control” function at some company and does lend itself to the type of confined
statistical analysis this assignment was aimed at. 
<br>
<br>
With the first question, we would intend to catch any variation that might depend on different
machinery used, such as if the different colored balls were made in separate sets of production
lines before being combined and mixed. 
<br>
<br>
However, if all balls came from the same production line before being colored (more likely),
there would only be small mass differences between the colors, if any. 
<br>
<br>
With the second question, we could imagine there might be some source of variation affecting
some piece of equipment having an influence on the mass, and that such variations might differ
in magnitude over time. 
<br>
<br>
We reasoned that if we sampled bags from a wide area, we would have a higher chance of
ending up with bags produced over a wider time period and thus be more likely of catching
such an error.
<br>
<br>

## Population & sampling
<br>
<br>
The population being studied is that of all bags of M&M Peanut balls, past, current and
future, that can be considered “comparable” for the purpose of this study.
<br>
<br>
By this we mean “coming from the same production conditions”, in the sense that we mainly
wanted to catch unintended production variations and not those due to deliberate changes
made to the production affecting ball mass. 
<br>
<br>
Lacking insight into the chocolate factory in question, for us this basically meant that we
assumed that no such changes had been made in the near past. 
<br>
<br>
Thus, our sample consisted of bags currently on the store shelves in our area and no vintage
candy was included.
<br>
<br>
Table XAZXZZXCZ shows an overview of the M&M bags used in the study. 


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv('CSVrawdata.csv')
bags = df.groupby('Location').sum()
bags.groupby('Mass')
```

```python
blue = df.loc[df['Color'] == 'Blue']
mass_blue= blue['Mass']
blue_mean = mass_blue.mean()
print("The mean is ", blue_mean, "grams for the blue M&M's")
plt.figure(figsize=(15,8))
plot = plt.hist(mass_blue, edgecolor='black', bins = 10, alpha = 0.75, color = '#00BFFF')
plt.title("Candy frequency for the Blue M&M's")
plt.xlabel('Milligrams')
plt.ylabel('Amount of candies')
plt.show()
```

```python
# basic sums
y_dot_dot_dot = df['Mass'].mean()
y_i_dot_dot = df.groupby('Color')['Mass'].mean()
y_dot_j_dot = df.groupby(['Store', 'Location', 'Number'])['Mass'].mean()
y_i_j_dot = df.groupby(['Store', 'Location', 'Number', 'Color'])['Mass'].mean()

# SS_A
balls_per_color = df.groupby('Color')['Mass'].count()
squares_A = (y_i_dot_dot - y_dot_dot_dot)**2
SS_A = (balls_per_color * squares_A).sum()

# SS_B
balls_per_bag = df.groupby(['Store', 'Location', 'Number'])['Mass'].count()
squares_B = (y_dot_j_dot - y_dot_dot_dot)**2
SS_B = (balls_per_bag * squares_B).sum()

# SS_AB
balls_per_bag_and_color = df.groupby(['Store', 'Location', 'Number', 'Color'])['Mass'].count()
squares_AB = (y_i_j_dot - y_i_dot_dot - y_dot_j_dot + y_dot_dot_dot)**2
SS_AB = (balls_per_bag_and_color * squares_AB).sum()

# SS_E
df_joined = df.join(y_i_j_dot, on=['Store', 'Location', 'Number', 'Color'], rsuffix='_other')
squares_E = (df_joined['Mass'] - df_joined['Mass_other']) ** 2
SS_E = squares_E.sum()

# dimensions
a = y_i_dot_dot.count()
b = y_dot_j_dot.count()

# degrees of freedom
a_deg_freedom = a-1
b_deg_freedom = b-1
ab_deg_freedom = (a-1) * (b-1)
e_deg_freedom = (balls_per_bag_and_color - 1).sum()

# mean squares
MS_A = SS_A / a_deg_freedom
MS_B = SS_B / b_deg_freedom
MS_AB = SS_AB / ab_deg_freedom
MS_E = SS_E / e_deg_freedom

# F-test statistics
F_A = MS_A/MS_E
F_B = MS_B/MS_E
F_AB = MS_AB/MS_E



print("\nSum of squares: ")

print(SS_A)
print(SS_B)
print(SS_AB)
print(SS_E)


print("\nDegrees of freedom: ")

print(a_deg_freedom)
print(b_deg_freedom)
print(ab_deg_freedom)
print(e_deg_freedom)


print("\nMean squares: ")

print(MS_A)
print(MS_B)
print(MS_AB)
print(MS_E)


print("\nF-test statistics: ")

print(F_A)
print(F_B)
print(F_AB)



print(balls_per_color)
```
